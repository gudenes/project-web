```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from pprint import pprint
from lxml import html
from lxml.html import fromstring
import urllib.request
from urllib.request import urlopen
import random
import re
#import scrapy
```

```python
url = "https://www.switchup.org/rankings/best-coding-bootcamps"
```

```python
html = requests.get(url).content
soup = bs(html, "lxml")
soup
```

```python
school = soup.find_all(["div",{"class":"ranking-item"}])[1].find_all("h3")
schools = [school[i].text.split(".")[1].strip() for i in range(1,51)]
```

```python
rating = soup.find_all("span",{"class":"ranking-item__rating--value"})
ratings = [rating[i].text.split("/")[0].strip() for i in range(len(rating))]
```

```python
rev = soup.find_all("span",{"class":"ranking-item__rating--count"})
reviews = [rev[i].text.split()[0] for i in range(len(rev))]
for i in range(len(reviews)):
    reviews[i] = reviews[i].replace("(","")
```

```python
details = soup.find_all("td")
details = [details[i].text.strip() for i in range(len(details))]

loc = [details[i] for i in range(0,len(details),3)]
courses = [details[i] for i in range(1,len(details),3)]
subject = [details[i] for i in range(2,len(details),3)]

```

```python
top_df=pd.DataFrame({"Schools":schools,"Ratings":ratings,"Reviews":reviews,"Locations":loc,"Courses":courses,"Subjects":subject})
top_df["Ratings"]=top_df["Ratings"].astype(float)
top_df.loc[0,"Reviews"] = top_df.loc[0,"Reviews"].replace(",","")
top_df["Reviews"]=top_df["Reviews"].astype(float)
```
